```python
#Dependencies
import pandas as pd
```

```python
#Get the path to your csv file
HofPymoli="Resources/purchase_data.csv"

#Read Purchase file and store into Pandas df
HofPymoli_df=pd.read_csv(HofPymoli)
HofPymoli_df.head(10)
```

```python
HofPymoli_df.columns
```

```python
#N of Players are equal to unique values count in columns SN (Players_Name)

N_of_Players=len(HofPymoli_df["SN"].unique())
N_of_Players
```

```python
#count of Number of Purchases
N_of_Purchases = HofPymoli_df["Purchase ID"].count()
N_of_Purchases
```

```python
#Total Revenue the Sum of single (Price) Purchased 
Total_Revenue = HofPymoli_df["Price"].sum()
Total_Revenue
```

```python
#Average Price the mean of Total single Prices
Average_Price =  round(HofPymoli_df["Price"].mean(),2)
Average_Price
```

```python
#N of Unique Items are equal to unique values count in column(Item Name)
N_unique_Items=len(HofPymoli_df["Item Name"].unique())
N_unique_Items
```

```python
#PURCHASING ANALYSIS
Data_Analysis=["N_unique_Items","Average_Price","N_of_Purchases","Total_Revenue"]

Data_Analysis_df=pd.DataFrame([{"No of Unique Items":N_unique_Items,
                                "$ Average Price":Average_Price,
                               "No of Purchases":N_of_Purchases,
                               "$ Total Revenue":Total_Revenue}
                              ])

Data_Analysis_df
```

```python
#Gender Demographics
#WE WANT THE UNIQUE VALUES IN SN THAT CONTAINS PLAYER NAMES SO WE DROP DUPLICATES,
#IN ORDER TO GROUP OUR DATA FRAME BY GENDER TO OBTAIN GENDER DEMOGRAPHICS
Players_df=HofPymoli_df.drop_duplicates(subset="SN")
Players_df
```

```python
#OUR NEW DF WILL CONTAIN ONLY GENDER INFORMATION 
Gender_N=["Female","Male","Other/Non-Disclosed"]
#Gender_df VALUE COUNTS
Gender_df=pd.DataFrame(Players_df["Gender"].value_counts())
New_Gender_df=Gender_df.rename(columns={"Gender":"Total of Players"})

New_Gender_df["Gender"]=Gender_N

Another_df=New_Gender_df.set_index("Gender")
Another_df

```

```python
#PERCENT CALCULATION USED ILOC TO GET THE VALUE IN CELLS AND DIVIDED BY N# OF PLAYERS 
Male_Percentage=round((Gender_df.iloc[0,0]/N_of_Players)*100,2)
Male_Percentage
```

```python
#PERCENT CALCULATION USED ILOC TO GET THE VALUE IN CELLS AND DIVIDED BY N# OF PLAYERS 
Female_Percentage=round((Gender_df.iloc[1,0]/N_of_Players)*100,2)
Female_Percentage
```

```python
#PERCENT CALCULATION USED ILOC TO GET THE VALUE IN CELLS AND DIVIDED BY N# OF PLAYERS 
Other_percentage=round((Gender_df.iloc[2,0]/N_of_Players)*100,2)
Other_percentage
```

```python
Percentage_of_Players=[Male_Percentage,Female_Percentage,Other_percentage]
Percentage_of_Players

```

```python
#INTRODUCED A NEW COLUMN Percentage_of_Players AND GIVED A NAME: Percent of Players %
Gender_df["Percent of Players %"]=Percentage_of_Players
Gender_df
```

```python
#CHANGED COLUMN NAME GENDER FOT TOTAL
#OBTAINED FINAL Gender Demographics
Gender_Demographics=Gender_df.rename(columns={"Gender":"Total Players"})
Gender_Demographics
```

```python
#I wanted to know the average Purchase done by Female, Male and others,
#I only need information of Gender and Price
New_df=HofPymoli_df[["Gender","Price"]]
New_df
```

```python
#Based on New_df info I can Group them get the Average Purchase (mean) done by Gender
Avg_Purchase = pd.DataFrame(round((New_df.groupby(["Gender"]).mean()),2))

#Chanched column name from Price to Avg Purchase Price
Avg_Purchase_df=Avg_Purchase.rename(columns={"Price":"Avg Purchase Price"})
Avg_Purchase_df
```

```python
#Based on New_df info I can Group them get the Purchase Sum done by Gender
Total_Purchase=pd.DataFrame(round((New_df.groupby(["Gender"]).sum()),2))

#Chanched column name from Price to Total Purchase
Total_Purchase_df=Total_Purchase.rename(columns={"Price":"Total Purchase"})
Total_Purchase_df
```

```python
#Based on Total_Purchase_df I located value Total Purchase female and got the Average Purchase male
Ave_Purchase_Female=round((Total_Purchase.iloc[0,0]/Gender_df.iloc[1,0]),2)
Ave_Purchase_Female
```

```python
#Based on Total_Purchase_df I located value Total Purchase male and got the Average Purchase male
Ave_Purchase_Male=round((Total_Purchase.iloc[1,0]/Gender_df.iloc[0,0]),2)
Ave_Purchase_Male
```

```python
#Based on Total_Purchase_df I located value Total Purchase others and got the Average Purchase others
Ave_Purchase_Others=round((Total_Purchase.iloc[2,0]/Gender_df.iloc[2,0]),2)
Ave_Purchase_Others
```

```python
#made a list to create a new DF Ave_Total_Person_df
Ave_Total_Person=[Ave_Purchase_Female,Ave_Purchase_Male,Ave_Purchase_Others]
Ave_Total_Person
```

```python

Ave_Total_df=pd.DataFrame({
    "Gender":["Female","Male","Other / Non-Disclosed"],
    "Avg Total Person $":[Ave_Purchase_Female,Ave_Purchase_Male,Ave_Purchase_Others]
})
Ave_Total_Person_df=Ave_Total_df.set_index("Gender")
Ave_Total_Person_df
```

```python
#JOINED DFs I used merge method 
PAvg_Analysis_df=pd.merge(Avg_Purchase_df,Ave_Total_Person_df, on="Gender", how="outer")
PAvg_Analysis_df
```

```python
#Total Purchasing Analysis Age
bins=[0,9,14,19,24,29,34,39,100]
group_labels=["0-9","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
HofPymoli_df["Age group"]=pd.cut(HofPymoli_df["Age"], bins, labels=group_labels)
HofPymoli_df
```

```python
#TOTAL PURCHASES VALUE BY AGE 
Age_Purchases=HofPymoli_df.groupby("Age group")
Total_Purchases_Age=round((Age_Purchases[["Price"]].sum()),2)
Total_Purchases_Age=Total_Purchases_Age.rename(columns={"Price":"$ Total Purchase Value"})
Total_Purchases_Age
```

```python
#AVERAGE PURCHASES BY AGE
Avg_Purchases_Age=round((Age_Purchases[["Price"]].mean()),2)

Avg_Purchases_Age=Avg_Purchases_Age.rename(columns={"Price":"$ Avg Purchases"})
Avg_Purchases_Age
```

```python
#COUNT OF PURCHASES BY AGE
#USED MY DATA FRAME Age_Purchases AND COUNTED THE PRICE COLUMNS
Ct_Purchases_Age=Age_Purchases[["Price"]].count()
```

```python
#CHANGED COLUMN NAME FOR Purchase Count
Ct_Purchases_Age=Ct_Purchases_Age.rename(columns={"Price":"Purchase Count"})
Ct_Purchases_Age
```

```python
#AVERAGE PURCHASE PER PERSON 
#I SELECTED ["Price","SN"] IN MY Age_Purchases FILE AND OBTAINED THE AVERAGE PER PERSON
Avg_Person=round((Age_Purchases[["Price","SN"]].mean()),2)
Avg_Person=Avg_Person.rename(columns={"Price":"$ Avg Purchase Price"})
Avg_Person
```

```python
#MERGE PURCHASE COUNT AND AVERAGE PURCHASE PER PERSON
Analysis_Age=pd.merge(Ct_Purchases_Age,Avg_Person,on="Age group", how="left" )
Analysis_Age
```

```python
# MERGE RESULT FROM PURCHASE COUNT AND AVG PURCHASE PER PERSON AND ADD TOTAL PURCHASE VALUE
Analysis=pd.merge(Analysis_Age,Total_Purchases_Age, on="Age group", how="left" )
Analysis
```

```python
#TOP SPENDERS
#FIND HOW MANY OCCURRENCES ARE IN PLAYERS (SN)
HofPymoli_df["SN"].value_counts()
```

```python
#GROUP BY PLAYERS 
Group_Player=HofPymoli_df.groupby(["SN"])
print(Group_Player)
```

```python
Group_Player.count().head()
```

```python
#SELECT DF BY PRICE AND COUNT
Total_count_Spenders=Group_Player["Price"].count()
Total_count_Spenders
Total_count_Spenders_df=pd.DataFrame({"Total Count":Total_count_Spenders})
Total_count_Spenders_df.head()
```

```python
#AVG PURCHASE PRICE BY SPENDER
```

```python
Avg_Spended= round((Group_Player["Price"].mean()),2)
Avg_Spended
Avg_Spended_df=pd.DataFrame({"$ Avg Spend":Avg_Spended})
Avg_Spended_df
```

```python
#TOTAL PURCHASE PRICE BY SPENDER
Purchase_Sum=Group_Player["Price"].sum()
Purchase_Sum
Purchase_Sum_df=pd.DataFrame({"$ Total Purchase Value":Purchase_Sum})
Purchase_Sum_df
```

```python
#MERGE TOTAL COUNT SPENDERS AND TOTAL PURCHASE PRICE BY SPENDER
Top_Spend=pd.merge(Total_count_Spenders_df,Avg_Spended_df, on="SN", how="left")
Top_Spenders=pd.merge(Top_Spend,Purchase_Sum_df, on="SN", how="left")
Top_Spenders
```

```python
#SORT DF BY TOTAL PURCHASE BY SPENDER AND TOTAL PURCHASE VALUE TO FIND WHO MADE MOST PURCHASES AND
#WHO SPENT THE MOST
#RESULT TOP SPENDERS
Top_Spenders_df = Top_Spenders.sort_values(
    ["Total Count", "$ Total Purchase Value"], ascending=False)
Top_Spenders_df.head()
```

```python
#MOST POPULAR ITEMS
columns=["Item ID","Item Name","Price"]

Pop_Items=HofPymoli_df[["Purchase ID","Item ID","Item Name","Price"]]
Pop_Items.head()
```

```python
Pop_Items_df=Pop_Items.groupby(["Item Name"]).count()
Pop_Items_df
```

```python
Total_Value_Items=Pop_Items.groupby(["Item Name"]).sum()
Total_Value_Items
```

```python
Sorted_Items=Pop_Items_df.sort_values(["Purchase ID","Item Name"],ascending=False)
Sorted_Items.head(10)
Sorted_Items=Sorted_Items.rename(columns={"Purchase ID":"Purchase count"})
Sorted_Items

Sorted_Items_df=pd.DataFrame(Sorted_Items)

Sorted_Items_df
```

```python
Item_df=HofPymoli_df[["Item Name","Price","Item ID"]]
Item_df
```

```python
#DROP DUPLICATES ON ITEM NAME AND WE HAVE 179 ITEMS 
Item_df=Item_df.drop_duplicates(subset="Item Name")
Item_df
```

```python
#USE DF Item_df SELECTED BY ITEMS AND MERGE WITH Sorted_Items_df

Popular=pd.merge(Sorted_Items_df,Item_df, on="Item Name", how="outer")
Popular.head()
```

```python
#MERGE Popular and Total_Value_Items
Most_Popular=pd.merge(Popular,Total_Value_Items, on="Item Name", how="outer")
Most_Popular
```

```python
#SELECT DF BY THE LIST ["Item Name","Item ID_y","Purchase count", "Price_y", "Price"]
Most_Popular_df=Most_Popular[["Item Name","Item ID_y","Purchase count", "Price_y", "Price"]]
Most_Popular_df
```

```python
#RENAME COLUMNS
Most_Popular_df=Most_Popular_df.rename(columns={"Item ID_y":"Item ID",
                                               "Price_y":"$ Price",
                                               "Price":"Total Price"})
Most_Popular_df
```
